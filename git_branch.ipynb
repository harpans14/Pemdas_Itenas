```python
import pandas as pd

data = {
    'Nama': ['John', 'Jane', 'Bob', 'Alice'],
    'Usia': [25, 35, 30, 28],
    'Gaji': [50000, 60000, 70000, 55000]
}

df = pd.DataFrame(data)
print("Data Awal:")
print(df)
```

No 1 Gunakan loop for dan fungsi lambda untuk menghitung gaji setiap karyawan setelah diberikan peningkatan sebesar 5% dari gaji saat ini.

```python

for i in range(len(df)):
    naik_5_persen = lambda gaji: gaji * 1.05
    df.loc[i, 'Gaji'] = naik_5_persen(df.loc[i, 'Gaji'])
```

No 2 Menampilkan DataFrame setelah perubahan + ringkasan

```python
print("\nData Setelah Kenaikan Gaji 5%:")
print(df)
print("Semua karyawan mendapatkan kenaikan gaji sebesar 5% dari gaji awal.")


```

No 3  Gunakan loop for lagi untuk mengevaluasi karyawan yang usianya di atas 30 tahun. Jika usia karyawan di atas 30, berikan peningkatan tambahan sebesar 2% dari gaji saat ini menggunakan fungsi lambda.


```python

for i in range(len(df)):
    if df.loc[i, 'Usia'] > 30:
        naik_2_persen = lambda gaji: gaji * 1.02
        df.loc[i, 'Gaji'] = naik_2_persen(df.loc[i, 'Gaji'])
```

No 4 Tampilkan DataFrame yang sudah diperbarui setelah peningkatan gaji tambahan dan berikan ringkasan hasilnya.

```python
print("\nData Setelah Kenaikan Tambahan 2% (Usia > 30):")
print(df)

print(" Semua karyawan mendapat kenaikan awal 5%.")
print(" Karyawan dengan usia di atas 30 tahun mendapat tambahan kenaikan 2%.")
```
